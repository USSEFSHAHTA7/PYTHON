```python
tasks = [{"task": "Quran", "completed": True},
         {"task": "Salah", "completed": True},
         {"task": "Study", "completed": False},
         {"task": "Exercise", "completed": True},
         {"task": "Sleep", "completed": False},
         {"task": "Visit jooooooo", "completed": True}]

def main():
    message = """
    1-add tasks to a list
    2-mark task as complete
    3-view tasks 
    4-Quit
    """
    
    while True:
        print(message)
        choice = input("Enter your choice: ")

        if choice == "1":
            add_task()
        elif choice == "2":
            complete_task()
        elif choice == "3":
            view_task(tasks)
        elif choice == "4":
            break
        else:
            print("Invalid choice, please enter choice from 1 to 4")

def add_task():
    task = input('Enter task: ')
    task_info = {"task": task, "completed": False}
    tasks.append(task_info)
    print("Task added to the list successfully")

def complete_task():
    incomplete_task = [] 
    for task in tasks:
        if task['completed'] == False:
            incomplete_task.append(task)
    if len(incomplete_task) == 0:
        print('No tasks to mark as complete')  
        return
    
    for i, task in enumerate(incomplete_task):
        print(f'{i+1}-{task}')
        print('_' * 30)

    try:
        task_number = int(input('Choose the task to complete: '))
        incomplete_task[task_number - 1]['completed'] = True
        print('Task marked as complete')
    except ValueError:
        print("Invalid input, please enter a number")

def view_task(tasks_list):
    if not tasks_list:
        print('No tasks to view')
        return
    for i, task in enumerate(tasks_list):
        if task["completed"]:
            status = '⭕' 
        else:
            status = '❌'
        print(f"{i + 1}.{task}={status}") 

if __name__ == "__main__":
    main()
```

```python

import os
def clear_screen():
  os.system('cls' if os.name=='nt' else 'clear')

class player:
    def __init__(self):
        self.name=''
        self.symbol=''
    def chose_name(self):
     while True:
      name=input('enter your name(letters only):')
      if name.isalpha():
        self.name=name
        break
      else:
        print('invalid name')
    def chose_symbol(self):
     while True:
      symbol=input(f'{self.name}, enter your symbol(X or O):')
      if symbol.isalpha() and (symbol=='x' or symbol=='o'):
        self.symbol=symbol.upper()
        break
      else:
        print('invalid symbol')

class menu:
  def display_main_menu(self):
    print('welcome to tic tac toe')
    print('1.start game')
    print('2.exit game')
    choice = input('enter your choice:')
    if choice == '1':
     return choice
  def display_end_game_menu(self):
    menu_text="""
    game over
    1.play again
    2.exit game
    """
    choice = input(menu_text)
    return choice

class board:
  def __init__(self):
    self.board=[]
    for i in range(1,10):
      self.board.append(str(i))
  def display_board(self):
    for i in range(0,9,3):
      print('|'.join(self.board[i:i+3]))
      if i<6:
        print('-'*5)
  def update_board(self,choice,symbol):
    if self.is_valid_move(choice): # Pass 'choice' to is_valid_move
      self.board[choice-1]=symbol
      return True
    else:
      return False
  def is_valid_move(self, choice): # Add 'choice' as a parameter
   return self.board[choice-1].isdigit()

  def reset_board(self):
    self.board=[]
    for i in range(1,10):
      self.board.append(str(i))

class game:
  def __init__(self):
    self.players=[player(),player()]
    self.board=board()
    self.menu=menu()
    self.current_play_index=0


  def start_game(self):
    choice = self.menu.display_main_menu() # Assign the return value of display_main_menu to choice
    if choice=='1':
      self.setup_players()
      self.play_game()
    else:
      self.Quit_game()

  def setup_players(self):
    for index,player in enumerate(self.players,start=1):
      print(f'player{index},enter your details')
      player.chose_name()
      player.chose_symbol()
      clear_screen()



  def play_game(self):
    while True:
      self.player_turn() # Changed from self.play_turn() to self.player_turn()
      if self.check_win() or self.check_draw():
       choice= self.menu.display_end_game_menu() # Use self.menu to access display_end_game_menu
       if choice=='1':
        self.reset_game()
       else:
        self.Quit_game()
      break

  def reset_game(self):
    self.board.reset_board()
    self.current_play_index=0
    self.play_game()


  def check_win(self):
    win_combinations=[[0,1,2],[3,4,5],[6,7,8],#row
                      [0,3,6],[1,4,7],[2,5,8],#column
                      [0,4,8],[2,4,6]]#diagonal
    for combination in win_combinations:
      if self.board.board[combination[0]]==self.board.board[combination[1]]==self.board.board[combination[2]]:
        return True
    return False

  def check_draw(self):
    for cell in self.board.board:
      if cell.isdigit():
        return False
    return True

  def player_turn(self):
    player=self.players[self.current_play_index]
    self.board.display_board()
    print(f'{player.name},turn ({player.symbol})')
    while True:
      try:
        choice=int(input('choose a cell (1,9):'))
        if 1<=choice<=9 and self.board.update_board(choice,player.symbol):
          break
        else:
          print('invalid move')
      except ValueError:
        print('please enter a number between 1 and 9')
    self.switch_player()

  def switch_player(self):
       self.current_play_index=1 - self.current_play_index

  def Quit_game(self):
    print('thank you for playing')

game=game()
game.start_game()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

class Person:
    def __init__(self, name, age):
        self.name = name
        self.age = age

class Student(Person):
    def __init__(self, name, age, grade, student_id):
        super().__init__(name, age)
        self.grade = grade
        self.student_id = student_id
        self.enrolled_classes = []
        self.extracurricular_activities = []
        self.clubs = []

    def join_club(self, club):
        self.clubs.append(club)
        club.members.append(self)
    
    def enroll_in_class(self, school_class):
        self.enrolled_classes.append(school_class)
        school_class.students.append(self)

class Teacher(Person):
    def __init__(self, name, age):
        super().__init__(name, age)
        self.classes_taught = []
        self.clubs_advised = []

    def advise_club(self, club):
        self.clubs_advised.append(club)
        club.advisor = self
    
    def teach_class(self, school_class):
        self.classes_taught.append(school_class)
        school_class.teacher = self

class Club:
    def __init__(self, club_name):
        self.club_name = club_name
        self.members = []
        self.advisor = None

class Class:
    def __init__(self, class_name):
        self.class_name = class_name
        self.students = []
        self.teacher = None

class School:
    def __init__(self):
        self.students = []
        self.teachers = []
        self.classes = []
        self.clubs = []

    def add_student(self, student):
        self.students.append(student)
        self.students = self.bubble_sort_students_by_id(self.students)

    def remove_student(self, student_name):
        self.students = [student for student in self.students if student.name != student_name]

    def add_teacher(self, teacher):
        self.teachers.append(teacher)

    def remove_teacher(self, teacher_name):
        self.teachers = [teacher for teacher in self.teachers if teacher.name != teacher_name]

    def add_class(self, school_class):
        self.classes.append(school_class)

    def add_club(self, club):
        self.clubs.append(club)

    def remove_club(self, club_name):
        self.clubs = [club for club in self.clubs if club.club_name != club_name]

    def get_every_student(self):
        return [student.name for student in self.students]

    def get_every_teacher(self):
        return [teacher.name for teacher in self.teachers]

    def get_every_class(self):
        return [school_class.class_name for school_class in self.classes]

    def get_every_club(self):
        return [club.club_name for club in self.clubs]

    def get_student_info_by_name(self, name):
        for student in self.students:
            if student.name == name:
                return {
                    "Name": student.name,
                    "Age": student.age,
                    "Grade": student.grade,
                    "ID": student.student_id,
                    "Enrolled Classes": [cls.class_name for cls in student.enrolled_classes],
                    "Clubs": [club.club_name for club in student.clubs]
                }
        return None

    def bubble_sort_students_by_id(self, students):
        n = len(students)
        for i in range(n):
            for j in range(0, n-i-1):
                if students[j].student_id > students[j+1].student_id:
                    students[j], students[j+1] = students[j+1], students[j]
        return students

    def binary_search_student_by_id(self, student_id):
        low, high = 0, len(self.students) - 1
        while low <= high:
            mid = (low + high) // 2
            if self.students[mid].student_id == student_id:
                return self.students[mid]
            elif self.students[mid].student_id < student_id:
                low = mid + 1
            else:
                high = mid - 1
        return None

    def visualize_school_data(self):
        data = {
            "Students": len(self.students),
            "Teachers": len(self.teachers),
            "Classes": len(self.classes),
            "Clubs": len(self.clubs)
        }
        sns.barplot(x=list(data.keys()), y=list(data.values()))
        plt.title("School Overview")
        plt.show()

def main():
    school = School()

    while True:
        print("\nSchool Management System")
        print("1. Add Student")
        print("2. Add Teacher")
        print("3. Add Class")
        print("4. Add Club")
        print("5. Enroll Student in Class")
        print("6. Assign Teacher to Class")
        print("7. Enroll Student in Club")
        print("8. Remove Student")
        print("9. Remove Teacher")
        print("10. Remove Club")
        print("11. Get All Student Names")
        print("12. Get All Teacher Names")
        print("13. Get All Class Names")
        print("14. Get All Club Names")
        print("15. Get Student Info by Name")
        print("16. Search Student by ID")
        print("17. Visualize School Data")
        print("18. Exit")

        choice = input("Enter your choice: ")

        if choice == '1':
            name = input("Enter student name: ")
            age = int(input("Enter student age: "))
            grade = int(input("Enter student grade: "))
            student_id = int(input("Enter student ID: "))
            student = Student(name, age, grade, student_id)
            school.add_student(student)
            print(f"Student {name} added.")

            club_name = input("Enter club name to join (or 'none' to skip): ")
            if club_name.lower() != 'none':
                club = next((c for c in school.clubs if c.club_name == club_name), None)
                if club:
                    student.join_club(club)
                    print(f"Student {name} joined club {club_name}.")
                else:
                    print("Club not found.")

            classes_enrolled = 0
            while classes_enrolled < 5:
                class_name = input("Enter class name to enroll: ")
                school_class = next((c for c in school.classes if c.class_name == class_name), None)
                if school_class:
                    student.enroll_in_class(school_class)
                    classes_enrolled += 1
                    print(f"Student {name} enrolled in class {class_name}.")
                else:
                    print("Class not found.")

        elif choice == '2':
            name = input("Enter teacher name: ")
            age = int(input("Enter teacher age: "))
            teacher = Teacher(name, age)
            school.add_teacher(teacher)
            print(f"Teacher {name} added.")

        elif choice == '3':
            class_name = input("Enter class name: ")
            school_class = Class(class_name)
            school.add_class(school_class)
            print(f"Class {class_name} added.")

        elif choice == '4':
            club_name = input("Enter club name: ")
            club = Club(club_name)
            school.add_club(club)
            print(f"Club {club_name} added.")

        elif choice == '5':
            student_name = input("Enter student name: ")
            class_name = input("Enter class name: ")
            student = next((s for s in school.students if s.name == student_name), None)
            school_class = next((c for c in school.classes if c.class_name == class_name), None)
            if student and school_class:
                student.enroll_in_class(school_class)
                print(f"Student {student_name} enrolled in class {class_name}.")
            else:
                print("Student or class not found.")

        elif choice == '6':
            teacher_name = input("Enter teacher name: ")
            class_name = input("Enter class name: ")
            teacher = next((t for t in school.teachers if t.name == teacher_name), None)
            school_class = next((c for c in school.classes if c.class_name == class_name), None)
            if teacher and school_class:
                teacher.teach_class(school_class)
                print(f"Teacher {teacher_name} assigned to class {class_name}.")
            else:
                print("Teacher or class not found.")

        elif choice == '7':
            student_name = input("Enter student name: ")
            club_name = input("Enter club name: ")
            student = next((s for s in school.students if s.name == student_name), None)
            club = next((c for c in school.clubs if c.club_name == club_name), None)
            if student and club:
                student.join_club(club)
                print(f"Student {student_name} joined club {club_name}.")
            else:
                print("Student or club not found.")

        elif choice == '8':
            student_name = input("Enter student name to remove: ")
            school.remove_student(student_name)
            print(f"Student {student_name} removed.")

        elif choice == '9':
            teacher_name = input("Enter teacher name to remove: ")
            school.remove_teacher(teacher_name)
            print(f"Teacher {teacher_name} removed.")

        elif choice == '10':
            club_name = input("Enter club name to remove: ")
            school.remove_club(club_name)
            print(f"Club {club_name} removed.")

        elif choice == '11':
            print("All student names:", school.get_every_student())

        elif choice == '12':
            print("All teacher names:", school.get_every_teacher())

        elif choice == '13':
            print("All class names:", school.get_every_class())

        elif choice == '14':
            print("All club names:", school.get_every_club())

        elif choice == '15':
            name = input("Enter student name to get info: ")
            student_info = school.get_student_info_by_name(name)
            if student_info:
                print("Student Info:", student_info)
            else:
                print("Student not found.")

        elif choice == '16':
            student_id = int(input("Enter student ID to search: "))
            student = school.binary_search_student_by_id(student_id)
            if student:
                print(f"Student found: {student.name}, ID: {student.student_id}")
            else:
                print("Student not found.")

        elif choice == '17':
            school.visualize_school_data()

        elif choice == '18':
            print("Exiting...")
            break

        else:
            print("Invalid choice, please try again.")

if __name__ == "__main__":
    main()
```
